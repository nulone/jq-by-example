## Example 1: Extract nested field

```
!jq-by-example \
  --input '{"user": {"name": "Alice", "age": 30}}' \
  --output '"Alice"' \
  --desc 'Extract user name'
```

## Example 2: Filter array

```
!jq-by-example \
  --input '[{"name": "Alice", "active": true}, {"name": "Bob", "active": false}]' \
  --output '[{"name": "Alice", "active": true}]' \
  --desc 'Keep only active users'
```
